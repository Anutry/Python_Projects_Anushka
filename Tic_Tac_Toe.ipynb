```python
import tkinter as tk
from tkinter import messagebox

class TicTacToe:
    def __init__(self):
    #to initialise our grid for tic_tac_toe
        self.current_player = "X"
        self.board = [["" for i in range(3)]for i in range(3)]
        self.window = tk.Tk()  
        self.buttons= [] #record of all buttons
        for i in range(3):
            row = []
            for j in range(3):
                button = tk.Button(self.window,text = "", width = "5", height = "5", command = lambda i=i ,j=j: self.make_move(i,j))
                button.grid(row = i, column = j)
                row.append(button)
            self.buttons.append(row)

    def make_move(self,row,column):
        #to record current_player's move
        if(self.board[row][column]==""):
            self.board[row][column] = self.current_player
            self.buttons[row][column].config(text=self.current_player)
            if(self.check_winner(self.current_player)):
                messagebox.showinfo(f"{self.current_player} won")
                self.window.quit()
                self.window.destroy()
            elif (self.is_board_full()):
                messagebox.showinfo("draw")
                self.window.quit()
                self.window.destroy()
            else:
                self.change_player()
            

    def check_winner(self,player):
            #to check if player is winner
            for i in range(3):
                if(self.board[i][0]==self.board[i][1]==self.board[i][2]==player):
                    return(True)
                elif(self.board[0][i]==self.board[1][i]==self.board[2][i]==player):
                    return(True)
                elif(self.board[0][0]==self.board[1][1]==self.board[2][2]==player):
                    return(True)
                elif(self.board[0][2]==self.board[1][1]==self.board[2][0]==player):
                    return(True)
            return(False)           
    

    def is_board_full(self):
        for i in range(3):
            for j in range(3):
                if self.board[i][j]=="":
                    return False
        return True
    def change_player(self):
        if(self.current_player == "X"):
            self.current_player = "O"
        else:
            self.current_player = "X"

    def run(self):
        self.window.mainloop()

game = TicTacToe()
game.run()




```
