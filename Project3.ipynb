```python
# SNAKE GAME YAYYY
import turtle
import time
import random

score=0
high_score=0
delay=0.2

window = turtle.Screen()
window.title("Snake Game")
window.bgcolor("lightblue")
window.setup(width=600, height=600)
window.tracer(0)

head = turtle.Turtle()
head.color("green")
head.shape("square")
head.direction = "Stop"
head.goto (0,0)
head.speed(0)
head.penup()

food = turtle.Turtle()
food.color("red")
food.shape("circle")
food.direction = "Stop"
food.speed(0)
food.penup()
food.goto(0,100)


sc = turtle.Turtle()
sc.hideturtle()
sc.penup()
sc.color("white")
sc.shape("square")
sc.goto(0,260)
sc.speed(0)
sc.write("Score:0 HighScore:0", align = "center" , font = ("Times New Roman",24,"bold"))


segments = []
def go_up():
    if head.direction!="down":
        head.direction = "up"
def go_down():
    if head.direction!="up":
        head.direction = "down"
def go_left():
    if head.direction!="right":
        head.direction = "left"
def go_right():
    if head.direction!="left":
        head.direction = "right"

def move():
    if head.direction=="up":
        y = head.ycor()
        head.sety(y+20)
    elif head.direction=="down":
        y = head.ycor()
        head.sety(y-20)
    elif head.direction=="left":
        x = head.xcor()
        head.setx(x-20)
    elif head.direction=="right":
        x = head.xcor()
        head.setx(x+20)
window.listen()
window.onkeypress(go_up,"Up")
window.onkeypress(go_down,"Down")
window.onkeypress(go_left,"Left")
window.onkeypress(go_right,"Right")

while True:
    window.update()

    #first see collision with border
    if(head.xcor()>290 or head.xcor()<-290 or head.ycor()>290 or head.ycor()<-290):
      
        time.sleep(1)

    
        head.goto(0,0)
        head.direction = "Stop"

        for segment in segments:
            segment.goto(1000,1000)
        segments.clear()
        score=0
        delay=0.1
        sc.clear()
        sc.write("Score : {} High Score : {} ".format(score,high_score), align = "center", font=("Times New Roman",24,"bold"))




    #collision with food
    if(head.distance(food)<20):
        x = random.randint(-290,290)
        y = random.randint(-290,290)
        food.goto(x,y)
        new_segment = turtle.Turtle()
        new_segment.color("grey")
        new_segment.speed(0)
        new_segment.penup()
        new_segment.shape("square")
        segments.append(new_segment)
        delay -= 0.01
        score+=10
        if score>high_score:
            high_score = score
        sc.clear()
        sc.write("Score : {} High Score : {} ".format(score,high_score), align = "center", font=("Times New Roman",24,"bold"))
    for i in range(len(segments)-1, -1, -1):
        x = segments[i-1].xcor() if i > 0 else head.xcor()
        y = segments[i-1].ycor() if i > 0 else head.ycor()
        segments[i].goto(x, y)

    move()

    #collision with body
    for segment in segments:
        if head.distance(segment)<20:
            time.sleep(1)
            head.goto(0,0)
            head.direction = "Stop"

            for segment in segments:
                segment.goto(1000,1000)
            segments.clear()
            score=0
            delay=0.1
            sc.clear()
            sc.write("Score : {} High Score : {} ".format(score,high_score), align = "center", font=("Times New Roman",24,"bold"))

    time.sleep(delay)

window.mainloop()
            

            


    
    
    
```
