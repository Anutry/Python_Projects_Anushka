```python
import tkinter as tk
from tkinter import messagebox



''' the functions we use are:
1) def __init__(self):
here, we will initialise the tkinter window /
name the tkinter window /
input the question data /
initialise current question index / 
make a question label /
make a frame for options : make frame in self.window and pack  /
we make buttons list /
then we make buttons for this options frames /
we make the next button /

2) def correct(self, index):
we check correctness of the option selected
this is the defining lambda function for buttons /

3) def load(self):
load next question /

4) def start(self):
load first question
call window mainloop

5) def next(next):
display next question /

6) how to call the quiz app?
-> initialize class
-> call start_quiz function
'''



class Quiz_app:
    def __init__(self):
        self.score = 0
        self.window = tk.Tk()
        self.window.title = "My Quiz"
        self.question_label = tk.Label(self.window,text = "")
        self.question_label.pack()
        self.current_idx = 0
        self.data = [
            {
                "question": "What is the capital of Japan?",
                "options": ["Tokyo", "Beijing", "Seoul", "Bangkok"],
                "correct_idx": 0
            },
            {
                "question": "Who wrote the novel 'Pride and Prejudice'?",
                "options": ["Jane Austen", "Emily Bronte", "Charlotte Bronte", "Louisa May Alcott"],
                "correct_idx": 0
            },
            {
                "question": "What is the chemical symbol for the element Gold?",
                "options": [ "Ag", "Au", "Cu", "Fe"],
                "correct_idx": 1
            },
        ]
        self.options_frame = tk.Frame(self.window)
        self.options_frame.pack()
        self.buttons = []

        for i in range(3):
            button = tk.Button(self.options_frame, text = "", width = 30, command = lambda i=i: self.correct(i))
            button.pack(pady=5)
            self.buttons.append(button)
        self.next = tk.Button (self.window, text = "Next Question", width = "30", command = lambda : self.next_question())
        self.next.pack(pady = 10)

    def correct(self,idx):
        question_data = self.data[self.current_idx]
        correct_idx = question_data["correct_idx"]
        if (idx == correct_idx) :
            self.score +=1
            messagebox.showinfo("Correct ;)")
        else:
            messagebox.showinfo("Incorrect :(")

    def next_question(self):
        self.current_idx += 1
        if (self.current_idx == len(self.data)):
            messagebox.showinfo("Quiz Over :D You scored",self.score)
            self.window.destroy()
        else:
            self.load()

    def load(self):
        question_data = self.data[self.current_idx]
        self.question_label.config(text = question_data["question"])
        options = question_data["options"]
        for i in range(3):
            self.buttons[i].config(text = options[i])

    def start(self):
        self.load()
        self.window.mainloop()

quiz = Quiz_app()
quiz.start()
```
